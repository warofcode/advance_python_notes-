# Question2: What do you mean by dynamically typed language ?


In programming, "typing" means checking the data types of variables.

Type checking is done in two Stages:

Static - Checking Data types before running the code.


Dynamic: Checking types while the code is running.

A statically typed language is one where the data types of variables are checked before the program is executed, typically during the compilation phase. In such languages, you need to explicitly declare the data type of each variable, and the compiler ensures that the operations performed on these variables are type-safe. If there is a mismatch between the declared data type and the operation being performed, the compiler will raise an error, preventing the program from running.

In simple terms, in statically typed languages like C or Java, you have to tell the compiler what type each variable is, and it makes sure you use them correctly throughout your code.

Example of Statically typed: Java , C++ ,C , Swift(iOS, macOS)

In a dynamically typed language, you don't need to declare the data type of a variable explicitly. Instead, the data type of a variable is determined automatically at runtime based on the value assigned to it. This means you can assign different types of values to the same variable without having to change its type declaration.

Example: Python ,Ruby , PHP ,Perl

Python is Both Compiled as well as Interpreted

.

"Python is an interpreted language", is the most common saying, which is also written in various books, but the hidden fact is Python is both compiled as well as an interpreted language. While running the code, Python generates a byte code internally, this byte code is then converted using a python virtual machine (p.v.m) to generate the output.

1. Pass statement

In genral pass keyword is used to fill up the empty blocks

The pass statement in Python is a null operation. It doesn't do anything when executed. It's used as a placeholder where syntactically required but no action is needed. It is often employed as a temporary placeholder when a statement is syntactically necessary but the program logic is not yet defined.

```python
#Example
for i in range(5):
    
        pass  # Placeholder, no action needed
```

2. Continue Statement

The continue statement is used to skip the rest of the code inside a loop for the current iteration, without terminating the loop entirely. It moves the control to the next iteration of the loop.

Syntax:

```python
for i in range(5):
    if i == 2:
        continue # skipping 2 (# if condition is true stop the current iteration)
    print(i)    
```

3. Break Statement

The break statement is used to exit a loop prematurely. When a break statement is encountered inside a loop, the loop is terminated immediately, and the program continues executing the code that follows the loop.

Syntax:

```python
for i in range(5):
    if i == 2:
        break # if condition is true terminate the entire loop
    print(i)
```
