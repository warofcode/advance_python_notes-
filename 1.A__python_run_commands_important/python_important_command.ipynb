# 2. I am not able to locate App data folder 

The AppData folder is a hidden folder in Windows. You may not see it by default in File Explorer. Here's how you can view it:

Open File Explorer.
    
Navigate to the directory: C:\Users\YourUsername.
    
Click on the "View" tab in the File Explorer menu.
    
Check the box that says "Hidden items". This will make hidden files and folders, including AppData, visible.
    
Now you should see the AppData folder in your user directory.
    
Once you've made hidden files and folders visible, you should be able to navigate to AppData -> Local -> Programs to find where Python is installed if you installed it using the standard installer. If you installed Python using Anaconda or another distribution, it may be installed in a different directory.


The AppData folder is hidden by default in Windows for several reasons:

System Integrity: The contents of the AppData folder are primarily used by applications to store settings, temporary files, cache, and other data. Hiding it helps prevent users from accidentally modifying or deleting critical files that could affect the functioning of their applications or the system.
User Experience: Windows aims to provide a clean and clutter-free user experience. Hiding system folders such as AppData reduces visual clutter in File Explorer and makes it easier for users to navigate their files.
Security: Some sensitive information, such as application settings and user data, might be stored in the AppData folder. Hiding it helps protect this information from being accessed or tampered with by unauthorized users.
While the AppData folder is hidden by default, users can still access it if needed by enabling the option to view hidden files and folders in File Explorer settings.

# 3. I have packages in Lib which came with python but when i install with pip it goes in site-packages

When you install packages using pip, they typically go into the site-packages directory. This is the default location for third-party Python packages installed using pip or other package managers.

The Lib directory, on the other hand, contains standard library modules and packages that come bundled with Python itself. These are the modules and packages developed and maintained by the Python Software Foundation and are available for all Python installations.

Here's a brief overview:

Lib directory: Contains the standard library modules and packages that come with Python.

site-packages directory: Contains third-party packages installed using tools like pip. These packages are not part of the standard library and are installed separately by the user.

This separation allows you to easily distinguish between the standard library modules provided by Python and third-party packages installed from external sources.
