```python
import math

class AREA_CALCULATOR:
    def __init__(self):
        pass
    
    def rectangle_area(self, length, width):
        return length * width
    
    def square_area(self, side):
        return side * side
    
    def circle_area(self, radius):
        return math.pi * radius * radius
    
    def triangle_area(self, base, height):
        return 0.5 * base * height
    
    def trapezoid_area(self, base1, base2, height):
        return 0.5 * (base1 + base2) * height
    
    def parallelogram_area(self, base, height):
        return base * height
    
    def ellipse_area(self, major_axis, minor_axis):
        return math.pi * major_axis * minor_axis
    
    def sector_area(self, radius, angle_degrees):
        angle_radians = math.radians(angle_degrees)
        return 0.5 * radius * radius * angle_radians
    
    def __str__(self):
        return "AREA_CALCULATOR class to calculate areas of different shapes"

# Example usage:
calc = AREA_CALCULATOR()

# Calculate areas
print("Rectangle Area:", calc.rectangle_area(5, 10))
print("Square Area:", calc.square_area(7))
print("Circle Area:", calc.circle_area(5))
print("Triangle Area:", calc.triangle_area(4, 6))
print("Trapezoid Area:", calc.trapezoid_area(3, 7, 4))
print("Parallelogram Area:", calc.parallelogram_area(5, 8))
print("Ellipse Area:", calc.ellipse_area(6, 3))
print("Sector Area:", calc.sector_area(8, 90))
```

# create separate class and constructor  for separate  area caculator

```python
import math

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width
    
class Square:
    def __init__(self, side):
        self.side = side
    
    def area(self):
        return self.side * self.side
    
class Circle:
    def __init__(self, radius):
        self.radius = radius
        self.pi = 3.14
    
    def area(self):
        return self.pi * self.radius * self.radius
    
class Triangle:
    def __init__(self, base, height):
        self.base = base
        self.height = height
    
    def area(self):
        return 0.5 * self.base * self.height
    
class Trapezoid:
    def __init__(self, base1, base2, height):
        self.base1 = base1
        self.base2 = base2
        self.height = height
    
    def area(self):
        return 0.5 * (self.base1 + self.base2) * self.height
    
class Parallelogram:
    def __init__(self, base, height):
        self.base = base
        self.height = height
    
    def area(self):
        return self.base * self.height
    
class Ellipse:
    def __init__(self, major_axis, minor_axis):
        self.major_axis = major_axis
        self.minor_axis = minor_axis
        self.pi = 3.14
    
    def area(self):
        return self.pi * self.major_axis * self.minor_axis
    
class Sector:
    def __init__(self, radius, angle_degrees):
        self.radius = radius
        self.angle_degrees = angle_degrees
        self.pi = 3.14
    
    def area(self):
        angle_radians = math.radians(self.angle_degrees)
        return 0.5 * self.radius * self.radius * angle_radians

# Example usage:
rect = Rectangle(length=5, width=10)
print("Rectangle Area:", rect.area())

sq = Square(side=7)
print("Square Area:", sq.area())

circ = Circle(radius=5)
print("Circle Area:", circ.area())

tri = Triangle(base=4, height=6)
print("Triangle Area:", tri.area())

trap = Trapezoid(base1=3, base2=7, height=4)
print("Trapezoid Area:", trap.area())

para = Parallelogram(base=5, height=8)
print("Parallelogram Area:", para.area())

ell = Ellipse(major_axis=6, minor_axis=3)
print("Ellipse Area:", ell.area())

sec = Sector(radius=8, angle_degrees=90)
print("Sector Area:", sec.area())
```
