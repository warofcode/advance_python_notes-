```python
import time
import requests

def check_server_health():
    while True:
        response = requests.get("http://yourserver.com/health")
        if response.status_code == 200:
            print("Server is healthy")
        else:
            print("Server is down")
        
        time.sleep(5)  # Check every 60 seconds

# Start monitoring the server
check_server_health()
```

```python
import tkinter as tk
from tkinter import messagebox
import requests

class ServerHealthChecker:
    def __init__(self, root):
        self.root = root
        self.root.title("Server Health Monitor")
        self.status_label = tk.Label(root, text="Checking server status...", font=("Helvetica", 16))
        self.status_label.pack(pady=20)
        self.check_server_health()

    def check_server_health(self):
        try:
            response = requests.get("http://codersinida.com/health")
            if response.status_code == 200:
                self.status_label.config(text="Server is healthy", fg="green")
            else:
                self.status_label.config(text="Server is down", fg="red")
        except requests.RequestException as e:
            self.status_label.config(text="Failed to connect", fg="orange")

        # Schedule the next check after 5 seconds (5000 milliseconds)
        self.root.after(5, self.check_server_health)

# Create the main window
root = tk.Tk()

# Create an instance of the ServerHealthChecker class
app = ServerHealthChecker(root)

# Run the Tkinter event loop
root.mainloop()
```
