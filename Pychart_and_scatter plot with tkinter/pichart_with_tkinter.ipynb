```python
# with put opps concept

import tkinter as tk
from tkinter import ttk
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import matplotlib.pyplot as plt

# Function to generate the pie chart
def plot_chart():
    # Data for the pie chart
    labels = ['A', 'B', 'C', 'D']
    sizes = [15, 30, 45, 10]
    colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
    explode = (0.1, 0, 0, 0)  # explode 1st slice

    # Create a pie chart
    fig = Figure(figsize=(6, 4), dpi=100)
    ax = fig.add_subplot(111)
    ax.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
           shadow=True, startangle=140)
    ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

    # Clear the frame before adding the new plot
    for widget in frame.winfo_children():
        if isinstance(widget, FigureCanvasTkAgg):
            widget.get_tk_widget().destroy()

    # Embed the plot in the Tkinter frame
    canvas = FigureCanvasTkAgg(fig, master=frame)
    canvas.draw()
    canvas.get_tk_widget().grid(row=1, column=0)

# Initialize the main application window
root = tk.Tk()
root.title("Pie Chart in Tkinter")

# Create a frame for the plot
frame = ttk.Frame(root, padding="10")
frame.grid(row=0, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))

# Create a button to generate the plot
plot_button = ttk.Button(frame, text="Generate Plot", command=plot_chart)
plot_button.grid(row=0, column=0)

# Run the Tkinter main loop
root.mainloop()
```

```python
# scatter chart 
import tkinter as tk
from tkinter import ttk
from matplotlib.figure import Figure
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import seaborn as sns
import matplotlib.pyplot as plt

class PyChartApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Seaborn Plot in Tkinter")

        # Create a frame for the plot
        self.frame = ttk.Frame(root, padding="10")
        self.frame.grid(row=0, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))

        # Create a button to generate the plot
        self.plot_button = ttk.Button(self.frame, text="Generate Plot", command=self.plot_chart)
        self.plot_button.grid(row=0, column=0)

    def plot_chart(self):
        # Generate a Seaborn plot
        data = sns.load_dataset('tips')
        fig = Figure(figsize=(6, 4), dpi=100)
        ax = fig.add_subplot(111)
        sns.scatterplot(x='total_bill', y='tip', hue='day', data=data, ax=ax)

        # Clear the frame before adding the new plot
        for widget in self.frame.winfo_children():
            if isinstance(widget, FigureCanvasTkAgg):
                widget.get_tk_widget().destroy()

        # Embed the plot in the Tkinter frame
        canvas = FigureCanvasTkAgg(fig, master=self.frame)
        canvas.draw()
        canvas.get_tk_widget().grid(row=1, column=0)

if __name__ == "__main__":
    root = tk.Tk()
    app = PyChartApp(root)
    root.mainloop()
```
